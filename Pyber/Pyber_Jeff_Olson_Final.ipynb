```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
#city_data_to_load = "data/city_data.csv"
#ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
City_Data_df = pd.read_csv("data/city_data.csv")
Ride_Data_df = pd.read_csv("data/ride_data.csv")
#print(City_Data_df.head())
#print(Ride_Data_df.head())

# Combine the data into a single dataset
city_and_ride_df = pd.merge(Ride_Data_df, City_Data_df, how='right', on="city")

# Display the data table for preview
city_and_ride_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
Urban_CR_df = city_and_ride_df.loc[city_and_ride_df["type"] == "Urban", :]

grouped_Urban_CR_df = Urban_CR_df.groupby(["city"])
#print("_____")

Urban_Rides_By_City = pd.DataFrame(grouped_Urban_CR_df["ride_id"].count())
#print(Rural_Rides_By_City)

Urban_Avg_Fare_By_City = pd.DataFrame(grouped_Urban_CR_df["fare"].mean())

#print(Rural_Avg_Fare_By_City)

Urban_merge_table_1 = pd.merge(Urban_Rides_By_City, Urban_Avg_Fare_By_City , on="city", how="right")
Urban_merge_table_1 


Urban_drivers_per_city_demographics = Urban_CR_df.loc[:, ["city", "driver_count"]]
Urban_drivers_per_city = pd.DataFrame(Urban_drivers_per_city_demographics.drop_duplicates())
#print(Urban_drivers_per_city)

Urban_merge_table_2 = pd.merge(Urban_merge_table_1 , Urban_drivers_per_city, on="city", how="right")
Urban_merge_table_2

Urban_Summary_df = Urban_merge_table_2.rename(columns={"city": "city",
                                        "ride_id": "Rides (per City)",
                                        "fare": "Urban",
                                        "driver_count": "Driver Density"})
Urban_Summary_df.head()

#Urban_Summary_df.dtypes
```

```python
Suburban_CR_df = city_and_ride_df.loc[city_and_ride_df["type"] == "Suburban", :]
grouped_Suburban_CR_df = Suburban_CR_df.groupby(["city"])
#print("_____")
Suburban_Rides_By_City = pd.DataFrame(grouped_Suburban_CR_df["ride_id"].count())
#print(Rural_Rides_By_City)

Suburban_Avg_Fare_By_City = pd.DataFrame(grouped_Suburban_CR_df["fare"].mean())
#print(Rural_Avg_Fare_By_City)

Suburban_merge_table_1 = pd.merge(Suburban_Rides_By_City, Suburban_Avg_Fare_By_City, on="city", how="right")
Suburban_merge_table_1 


Suburban_drivers_per_city_demographics = Suburban_CR_df.loc[:, ["city", "driver_count"]]
Suburban_drivers_per_city = pd.DataFrame(Suburban_drivers_per_city_demographics.drop_duplicates())

Suburban_merge_table_2 = pd.merge(Suburban_merge_table_1 , Suburban_drivers_per_city, on="city", how="right")
Suburban_merge_table_2

Suburban_summary_df = Suburban_merge_table_2.rename(columns={"city": "city",
                                        "ride_id": "Rides (per City)",
                                        "fare": "Suburban",
                                        "driver_count": "Driver Density",
                                         })
Suburban_summary_df.head()
```

```python
Rural_CR_df = city_and_ride_df.loc[city_and_ride_df["type"] == "Rural", :]
grouped_Rural_CR_df = Rural_CR_df.groupby(["city"])
#print("_____")
Rural_Rides_By_City = pd.DataFrame(grouped_Rural_CR_df["ride_id"].count())
#print(Rural_Rides_By_City)

Rural_Avg_Fare_By_City = pd.DataFrame(grouped_Rural_CR_df["fare"].mean())
#print(Rural_Avg_Fare_By_City)

Rural_merge_table_1 = pd.merge(Rural_Rides_By_City, Rural_Avg_Fare_By_City, on="city", how="right")
Rural_merge_table_1 


drivers_per_city_demographics = Rural_CR_df.loc[:, ["city", "driver_count"]]
drivers_per_city = pd.DataFrame(drivers_per_city_demographics.drop_duplicates())

Rural_merge_table_2 = pd.merge(Rural_merge_table_1 , drivers_per_city, on="city", how="right")
Rural_merge_table_2

#Rural_Summary_df =pd.DataFrame([ {"city": "city", "ride_id":Rural_Rides_By_City,"fare":Rural_Avg_Fare_By_City,
#         "driver_count":drivers_per_city} ])
#Rural_Summary_df["fare"] = Rural_Summary_df["fare"].map("${:.2f}".format)

Rural_Summary_df = Rural_merge_table_2.rename(columns={"city": "city",
                                        "ride_id": "Rides (per City)",
                                        "fare": "Rural",
                                        "driver_count": "Driver Density",
                                         })
Rural_Summary_df.head()
```

```python
# Build the scatter plots for each Urban city types
x_axis_Urban = Urban_Summary_df["Rides (per City)"]
#print(x_axis)

y_data_Urban= Urban_Summary_df["Urban"]
#print(y_data_city_plus_ride_df)

bubbles_Urban=Urban_Summary_df["Driver Density"]

plt.scatter(x_axis_Urban, y_data_Urban, marker="o", facecolors="lightcoral", edgecolors="black",
            s=7*bubbles_Urban, alpha=0.5)
plt.title("Pyber Ride Sharing Data 2018  " + 
          " (Bubble size ~ Drivers per City)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Urban Fare ($)")
labels = ['Urban', 'Suburban', 'Rural']

# Build the scatter plots for each Suburban city types

x_axis_Suburban = Suburban_summary_df["Rides (per City)"]
#print(x_axis)

y_data_Suburban= Suburban_summary_df["Suburban"]
#print(y_data_city_plus_ride_df)
bubbles_Suburban=Suburban_summary_df["Driver Density"]

plt.scatter(x_axis_Suburban, y_data_Suburban, marker="o", facecolors="lightskyblue", edgecolors="black",
            s=7*bubbles_Suburban, alpha=0.5)
plt.ylabel("Average Suburban Fare ($)")


# Build the scatter plots for each Suburban city types
x_axis_Rural = Rural_Summary_df["Rides (per City)"]
#print(x_axis)

y_data_Rural= Rural_Summary_df["Rural"]
#print(y_data_city_plus_ride_df)
bubbles_Rural=Rural_Summary_df["Driver Density"]

# Build the scatter plots for each city types
plt.scatter(x_axis_Rural, y_data_Rural, marker="o", facecolors="gold", edgecolors="black",
            s=7*bubbles_Rural, alpha=0.5)

plt.ylabel("Average Rural Fare ($)")

plt.grid()
plt.legend(loc='best')
# Save the plot and display it
plt.savefig("../Images/Fares_By_Rides_Bubble_Plot_JFO.png")
plt.show()
```

## Total Fares by City Type

```python
%matplotlib inline
```

```python
# Calculate Type Percents
total_fare=city_and_ride_df["fare"].sum()

total_urban_fare = Urban_CR_df["fare"].sum()
total_suburban_fare = Suburban_CR_df["fare"].sum()
total_rural_fare = Rural_CR_df["fare"].sum()

urban_fare_pct = total_urban_fare/total_fare
suburban_fare_pct = total_suburban_fare/total_fare
rural_fare_pct = total_rural_fare/total_fare

# Build Pie Chart
# Labels for the sections of our pie chart
labels = ["Urban","Rural", "Suburban"]

# The values of each section of the pie chart
sizes = [urban_fare_pct,  rural_fare_pct, suburban_fare_pct,]

# The colors of each section of the pie chart
colors = ["lightcoral", "gold","lightskyblue" ]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0.3, 0, 0)

plt.pie(sizes,explode =explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=270)
plt.title("% Total Fares By City Type")

# Save Figure
plt.savefig("../Images/Total_Fares_By_City_Type_Pie_JFO.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
total_ride_count=city_and_ride_df["ride_id"].count()

urban_ride_count=Urban_CR_df["ride_id"].count()
suburban_ride_count=Suburban_CR_df["ride_id"].count()
rural_ride_count=Rural_CR_df["ride_id"].count()

Urban_ride_pct=urban_ride_count/total_ride_count
suburban_ride_pct=suburban_ride_count/total_ride_count
rural_ride_pct=rural_ride_count/total_ride_count

# Build Pie Chart
# Labels for the sections of our pie chart
labels = ["Urban", "Rural","Suburban"]

# The values of each section of the pie chart
sizes = [Urban_ride_pct, rural_ride_pct, suburban_ride_pct]

# The colors of each section of the pie chart
colors = ["lightcoral", "gold","lightskyblue" ]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0.3, 0, 0)

plt.pie(sizes,explode =explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=270)
plt.title("% Total Rides By City Type")

# Save Figure
plt.savefig("../Images/Total_Rides_By_City_Type_Pie_JFO.png")
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
total_drivers=city_and_ride_df["driver_count"].sum()

total_urban_drivers=Urban_CR_df["driver_count"].sum()
total_suburban_drivers=Suburban_CR_df["driver_count"].sum()
total_rural_drivers=Rural_CR_df["driver_count"].sum()


urban_driver_pct=total_urban_drivers/total_drivers
suburban_driver_pct=total_suburban_drivers/total_drivers
rural_driver_pct=total_rural_drivers/total_drivers

# Build Pie Chart
# Labels for the sections of our pie chart
labels = ["Urban","Rural","Suburban"]

# The values of each section of the pie chart
sizes = [urban_driver_pct, rural_driver_pct, suburban_driver_pct]

# The colors of each section of the pie chart
colors = ["lightcoral", "gold","lightskyblue"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0.3, 0, 0)

plt.pie(sizes,explode =explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=240)
plt.title("% Total Drivers By City Type")

# Save Figure
plt.savefig("../Images/Total_Drivers_By_City_Type_Pie_JFO.png")

# Show Figure
plt.show()
```
